# Importing Modules

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('tagsets')
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk import pos_tag
```

# Text

```python
text = """Modern civilization is based upon the use of machines. For doing their workday in and day out the machines need a constant supply
of energy without which they would come to a halt. Our transport systems-diesel-run railways, steamships, motor vehicles, and aeroplanes-all
constantly need natural fuel to keep on moving. In the last three hundred years, many human efforts have been spent in search of the sources
of energy-coal, petroleum, and power generated from water–which will maintain the present rhythm of civilization unchecked.Not only that;
but as the population of the world increases, the need for energy will also increase. But there is a real fear that the world’s reserves
for energy sources of petroleum, coal, and hydel power are gradually being exhausted. That is why new and unconventional sources of energy
like nuclear and solar energy need to be developed.While the advanced countries in America and Europe have the wealth and scientific
know-how to produce solar and nuclear energy on a commercial scale, the poorer Asiatic countries like India, Pakistan, and Bangladesh may 
develop an energy source by bio-gas-developing machines. But the point is that research in the field of the development of new energy 
sources must be carried on further and expedited."""
```

# Tokenization

```python
words = word_tokenize(text)
words
```

```python
sentences=sent_tokenize(text)
sentences
```

# StepWords

```python
filter_list=[]
stop_words=set(stopwords.words("english"))
for word in words:
    if word.casefold() not in stop_words:
         filter_list.append(word)

filter_list
```

# Stemming

```python
stemr=PorterStemmer()
stemmed_words=[stemr.stem(word) for word in words]
stemmed_words
```

# PartsOfSpeach

```python
partsofspeach=pos_tag(words)
partsofspeach
```

```python
nltk.help.upenn_tagset()
```

# Lemmatizing

```python
lemmatizer=WordNetLemmatizer()
lemmatizedWords=[lemmatizer.lemmatize(word) for word in words]
lemmatizedWords
```

# Chunking

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"
chunk_parser = nltk.RegexpParser(grammar)
phrases=chunk_parser.parse(partsofspeach)
phrases.draw()
```

# Chinking

```python
grammar = """
... Chunk: {<.*>+}
...        }<JJ>{"""
chunk_parser2 = nltk.RegexpParser(grammar)
phrases2=chunk_parser2.parse(partsofspeach)
phrases.draw()

```
