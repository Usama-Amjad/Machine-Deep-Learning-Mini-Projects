#### In this course project, we will build a regression model using the deep learning Keras library, and then we will experiment with increasing the number of training epochs and changing number of hidden layers and we will see how changing these parameters impacts the performance of the model

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

# Part A

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
##Creating a model
def regression_model(columns):
    model= Sequential()
    model.add(Dense(10, activation='relu', input_shape=(columns-1,)))
    model.add(Dense(1)) 
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
errors1=[]
# Splitting data into X and y
X=concrete_data.drop(['Strength'],axis=1)
y=concrete_data.Strength

# Determining MSE and repeating the process and appending the mSE in errors
for cycles in range(50):
    # Splitting Data
    x_train, x_test, y_train, y_test = train_test_split(X, 
                                                        y, 
                                                        test_size=0.3,
                                                       random_state=23)
    model=regression_model(concrete_data.shape[1])
    # Fitting and predicting
    model.fit(x_train,y_train,epochs=50)
    y_pred=model.predict(x_test)
    mse= mean_squared_error(y_test,y_pred)
    errors1.append(mse)
```

```python
std1=np.std(errors1)
mean1=np.mean(errors1)
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE)".format(std1,mean1))
```

# Part B

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
# Normalizing Data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
errors2=[]
# Repeating same Process as in A but with normalized Data
for cycles in range(50):
    x_train2, x_test2, y_train2, y_test2 = train_test_split(predictors_norm, 
                                                        target, 
                                                        test_size=0.3,
                                                       random_state=23)
    model=regression_model(concrete_data.shape[1])
    model.fit(x_train2,y_train2,epochs=50)
    y_pred2=model.predict(x_test2)
    mse= mean_squared_error(y_test2,y_pred2)
    errors2.append(mse)
    
```

```python
std2=np.std(errors2)
mean2=np.mean(errors2)
```

```python
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE)".format(std1,mean1))
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE)".format(std2,mean2))
```

# Part C

```python
errors3=[]
# Repeating the same process as B but with epochs=100
for cycles in range(50):
    x_train3, x_test3, y_train3, y_test3 = train_test_split(predictors_norm, 
                                                        target, 
                                                        test_size=0.3,
                                                       random_state=23)
    model=regression_model(concrete_data.shape[1])
    model.fit(x_train3,y_train3,epochs=100)
    y_pred3=model.predict(x_test3)
    mse= mean_squared_error(y_test3,y_pred3)
    errors3.append(mse)
std3=np.std(errors3)
mean3=np.mean(errors3)
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE) of PartB".format(std2,mean2))
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE) of PartC".format(std3,mean3))
```

# Part D

```python
# Creating a new model with additional layers
def regression_model2(columns):
    model= Sequential()
    model.add(Dense(10, activation='relu', input_shape=(columns-1,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1)) 
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
model=regression_model2(concrete_data.shape[1])
errors4=[]
# Repeating the same process as B but with different model
for cycles in range(50):
    x_train4, x_test4, y_train4, y_test4 = train_test_split(predictors_norm, 
                                                        target, 
                                                        test_size=0.3,
                                                       random_state=23)
    model=regression_model(concrete_data.shape[1])
    model.fit(x_train4,y_train4,epochs=50)
    y_pred4=model.predict(x_test4)
    mse= mean_squared_error(y_test4,y_pred4)
    errors4.append(mse)
```

```python
std4=np.std(errors4)
mean4=np.mean(errors4)
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE) of PartB".format(std2,mean2))
print("Standard Deviation:{0:.3f} and Mean:{0:.3f} of Mean Squared Errors(MSE) of PartD".format(std4,mean4))
```
